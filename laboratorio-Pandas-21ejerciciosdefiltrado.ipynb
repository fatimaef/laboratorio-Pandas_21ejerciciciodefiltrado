#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
titles_90min = df_netflix_producciones[df_netflix_producciones['duration'] == '90 min'][['title', 'duration']].reset_index(drop=True)
print(titles_90min)
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
titles_india_after_2018 = df_netflix_producciones[(df_netflix_producciones['country'] == 'India') & (df_netflix_producciones['release_year'] > 2018)][['title', 'release_year']]
titles_india_after_2018.reset_index(drop=True, inplace=True)
print(titles_india_after_2018)
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
titles_ratings = df_netflix_producciones[df_netflix_producciones['rating'].isin(['TV-14', 'R'])][['title', 'rating']]
titles_ratings.reset_index(drop=True, inplace=True)
print(titles_ratings)
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
directors_filter = ['Christopher Nolan', 'Martin Scorsese']
titles_directors = df_netflix_producciones[df_netflix_producciones['director'].isin(directors_filter)][['title', 'director']]
titles_directors.reset_index(drop=True, inplace=True)
directors_count = titles_directors['director'].value_counts()
print(directors_count)
```

#### Ejercicio 17: Encuentra los 10 títulos con las puntuaciones más altas en IMDb.

#### Ejercicio 18: Encuentra los 5 títulos con las puntuaciones más bajas en IMDb.

#### Ejercicio 19: Filtra todos los títulos que no tienen un director asignado (valores nulos en la columna 'director').

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.
